```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Flatten,Dense,Input
from tensorflow.keras import Model
from tensorflow.keras.callbacks import Callback

# Prepare the datasets
(x_train,y_train),(x_test,y_test)=mnist.load_data()
x_train,x_test=x_train/255.0,x_test/255.0

# Create Model
x_input=Input(shape=(28,28))
x=Flatten(input_shape=(28,28))(x_input)
x=Dense(512,activation='relu')(x)
output_=Dense(10,activation='softmax')(x)
model=Model(inputs=x_input,outputs=output_)

# Create a callback to stop
class Stop_at_acc_99_cb(Callback):
    def on_epoch_end(self,epoch,logs={}):
        if(logs.get('accuracy')>0.99):
            print("\n Reached 99% accuracy so cancelling training!")
            self.model.stop_training=True

callbacks = Stop_at_acc_99_cb()


# Compile model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

# fit the model
model.fit(x_train, y_train, epochs=10, callbacks=[callbacks])
```
